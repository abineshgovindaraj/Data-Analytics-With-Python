## Scraping Website

```python
url="https://www.flipkart.com/search?q=redmi+note+10+pro&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=redmi+note+10+pro%7CMobiles&requestId=48fe63fc-0196-44cf-8e56-a963b69012c3&as-searchtext=Red"
```

## Response

```python
response=requests.get(url)
```

```python
response
```

## Parsing

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
names=soup.select('._4rR01T')
```

```python
names
```

```python
mobile_names=[]
```

```python
for item in names:
    mobile_names.append(item.text)
    mobile_names
```

```python
mobile_names
```

```python
price=soup.select('._1_WHN1')
```

```python
price
```

```python
mobile_price=[]
```

```python
for item in price:
    mobile_price.append(item.text)
    mobile_price
```

```python
 mobile_price
```

```python
reviews=soup.select('._13vcmD+ span')
```

```python
reviews
```

```python
mobile_reviews=[]
```

```python
for item in reviews:
    mobile_reviews.append(item.text)
    mobile_reviews
```

```python
mobile_reviews
```
